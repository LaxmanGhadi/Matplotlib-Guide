```python
import matplotlib.pyplot as plt
import numpy as np4
import pandas as pd
# get csv file from files
data_frame = pd.read_csv('/content/loan.csv')

# getting credit score and income of only male candidates
credit_score = data_frame[data_frame['gender']=='Male']['credit_score']
income = data_frame[data_frame['gender']=='Male']['income']

# using 's' to set the size and color to set the color of plotpoints
plt.scatter(income,credit_score,s = 100,linewidth = 1, color='green')


# getting credit score and income of only female candidates
credit_score_f = data_frame[data_frame['gender']=='Female']['credit_score']
income_f = data_frame[data_frame['gender']=='Female']['income']

# using 's' to set the size and color to set the color of plotpoints
plt.scatter(income_f,credit_score_f, s = 100,linewidth = 1,color='red')


#add a title to the graph
plt.title('income vs credit score')
# label the x axis
plt.xlabel('Income')
# label the y axis
plt.ylabel('credit score')

#adding legends to tell the viewer which color denotes what category
plt.legend(['Male','Female'], loc = 'lower right')
plt.tight_layout()
plt.show()
```

```python
# marker is the mark that is used to plot a point on graph
# You can customize the way your marker looks which can become useful to convey different insights in your data and convey useful information
# when you are plotting a graph which represents data in points like scatter plot
# It takes the fillowing arguements
# s -> s can be an integer which decides the size of marker
# color -> color of marker can be a name of color string
# type -> can be a string eg '*' -> will give a star

credit_score = data_frame[data_frame['gender']=='Male']['credit_score'][:10]
income = data_frame[data_frame['gender']=='Male']['income'][:10]
plt.scatter(income,credit_score,s = 'Large',linewidth = 1, color='green')

credit_score = data_frame[data_frame['gender']=='Male']['credit_score'][10:20]
income = data_frame[data_frame['gender']=='Male']['income'][10:20]
plt.scatter(income,credit_score,s = 200,linewidth = 1, color='red')
```

```python
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = numpy.random.binomial(1,.9,size=1000)
predicted = numpy.random.binomial(1,.9,size=1000)

confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix , display_labels = [False,True])
cm_display.plot()
plt.show()
```
